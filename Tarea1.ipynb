## Punto 2.0.2 Maximos. 

```python
import numpy as np
import matplotlib.pyplot as plt

def datos(archivo):
    datos= np.loadtxt(archivo)
    x =  datos[0 :,0]
    y = datos[0:,1]
    return (x,y)
```

```python
x, y = datos('Datospunto1.txt')
h= len(y)
listay= np.empty(0)
listax=np.empty(0)
for i in range(h):
    if i == 0: 
        if y[i]-y[i+1] > 0: 
            listay = np.append(listay, y[i])
            listax = np.append(listax, x[i])
    elif i > 0 and i < h-1: 
        if y[i] - y[i-1] > 0 and y[i] - y[i+1] > 0: 
            listay = np.append(listay, y[i])
            listax = np.append(listax, x[i])
    elif i == h-1:
        if y[i]-y[i-1] > 0:
            listay = np.append(listay, y[i])
            listax = np.append(listax, x[i])
plt.plot(x,y, color = 'blue')
plt.scatter(listax, listay, color='orange', marker= 'o')
plt.title('Maximos Locales')
plt.xlabel('eje X')
plt.ylabel('eje Y')
plt.show
```

## Punto 6. Derivacion. 

```python
import numpy as np 
import matplotlib.pyplot as plt
#a Definir una discretizacion en los ejes X y Y con 25 puntos en casa eje. 
n = 25
x = np.linspace(-4,4,n)
y = np.linspace(-4,4,n)
X,Y = np.meshgrid(x,y)

#b Definir la funcion potencial del flujo. 
h= 0.001 #pasos de la derivada.
V = 2 #cm/s
R = 2 #cm
phi = V*X*(1- (R**2/(X**2 + Y**2)))
# definimos los puntos dentro del circulo con la formula de la ecuacion general de un circulo 
circulo = (X**2 + Y**2) <= R**2
phi[circulo] = 0

v_x = np.zeros_like(phi)
v_y = np.zeros_like(phi)

for i in range(1,n-1):
    for j in range(1, n-1):
        v_x[i,j]= (phi[i, j + 1] - phi[i, j - 1]) / (2 * h)
        v_y[i,j] = -(phi[i + 1, j] - phi[i - 1, j]) / (2 * h)

v_x[circulo]= 0
v_y[circulo] = 0



plt.figure(figsize=(6, 6))
plt.quiver(X, Y, v_x, v_y)
plt.xlim([-4, 4])
plt.ylim([-4, 4])
plt.xlabel('x (cm)')
plt.ylabel('y (cm)')
plt.title('Campo de velocidades cerca de la superficie de un cilindro')
plt.grid(True)
plt.gca().set_aspect('equal')  # Mantener la proporcin de los ejes para un crculo correcto
plt.show()


```
