```python
import random

def es_un_par(dados):

    conteos = [dados.count(i) for i in range(1, 7)]
    return conteos.count(2) == 1 and conteos.count(1) == 3

def son_dos_pares(dados):

    conteos = [dados.count(i) for i in range(1, 7)]
    return conteos.count(2) == 2 and conteos.count(1) == 1

def lanzar_dados():
    return [random.randint(1, 6) for _ in range(5)]

N = 10**5

exitos_un_par = 0
exitos_dos_pares = 0

for _ in range(N):
    dados = lanzar_dados()
    if es_un_par(dados):
        exitos_un_par += 1
    if son_dos_pares(dados):
        exitos_dos_pares += 1

probabilidad_un_par = exitos_un_par / N
probabilidad_dos_pares = exitos_dos_pares / N

print(f"Probabilidad experimental de obtener exactamente 1 par: {probabilidad_un_par:.4f}")
print(f"Probabilidad experimental de obtener exactamente 2 pares: {probabilidad_dos_pares:.4f}")
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt

k_B = 1.0
N = 1000
epsilon_0 = 1
epsilon_1 = 2

def omega(N, n0):
    n1 = N - n0

    return math.factorial(N) / (math.factorial(n0) * math.factorial(n1))

def entropia(N, n0):
    n1 = N - n0
    omega_val = omega(N, n0)
    # Entropía S = k_B * ln(Ω)
    return k_B * np.log(omega_val)

def entropia_x(N, x):
    # S = -k_B * N * [x * ln(x) + (1 - x) * ln(1 - x)]
    return -k_B * N * (x * np.log(x) + (1 - x) * np.log(1 - x))

def graficar_entropia():
    x_vals = np.linspace(0.01, 0.99, 100)
    S_vals = entropia_x(N, x_vals) / (k_B * N)

    plt.plot(x_vals, S_vals)
    plt.xlabel("Fracción x")
    plt.ylabel("S(N, x) / (k_B N)")
    plt.title("Entropía vs Fracción de partículas en el nivel ϵ1")
    plt.grid(True)
    plt.show()

def proporcion_x(T, delta_E, k_B=1.0):

    T = np.array(T)

    if np.any(T <= 0):
        return np.zeros_like(T)

    x_bajas = T < 1e-5
    x_altas = T > 1e5

    exp_term = delta_E / (k_B * T)

    exp_term = np.clip(exp_term, -700, 700)

    x_intermedias = 1 / (1 + np.exp(exp_term))

    x = np.where(x_bajas, 0, np.where(x_altas, 0.5, x_intermedias))

    return x

def limites_x():
    # Para T -> 0
    x_0 = proporcion_x([1e-10], epsilon_1 - epsilon_0)[0]
    # Para T -> ∞
    x_inf = proporcion_x([1e10], epsilon_1 - epsilon_0)[0]
    return x_0, x_inf

def cambio_entropia_expansion(V1, V2, N, k_B):
    # ΔS = N * k_B * ln(V2 / V1)
    return N * k_B * np.log(V2 / V1)

def resolver_problema():
    n0 = N // 2
    omega_val = omega(N, n0)
    print(f"(a) Número de configuraciones posibles Ω(N, n0) = {omega_val}")

    S_val = entropia(N, n0)
    print(f"(b) Entropía S(N, n0) = {S_val}")

    x = 0.5
    S_x = entropia_x(N, x)
    print(f"(c) Entropía S(N, x) = {S_x}")

    print("(d) Graficando la entropía en función de x...")
    graficar_entropia()

    T_vals = np.linspace(0.1, 10, 100)
    x_vals_T = proporcion_x(T_vals, epsilon_1 - epsilon_0)

    plt.plot(T_vals, x_vals_T)
    plt.xlabel("Temperatura T")
    plt.ylabel("Proporción de partículas x(T)")
    plt.title("Proporción de partículas x(T) vs Temperatura T")
    plt.grid(True)
    plt.show()

    x_0, x_inf = limites_x()
    print(f"(f) Limite T -> 0: x(T) = {x_0}")
    print(f"Limite T -> ∞: x(T) = {x_inf}")

    S_inf = k_B * N * np.log(2)
    print(f"S(T) a T -> ∞: S(T) = {S_inf}")

    V1 = 1.0  # Volumen inicial
    V2 = 2.0  # Volumen final
    delta_S = cambio_entropia_expansion(V1, V2, N, k_B)
    print(f"(g) Cambio de entropía en expansión isotérmica: ΔS = {delta_S}")

resolver_problema()
```

```python
import numpy as np

T = np.array([[0.8, 0.2],
              [0.2, 0.8]])

E = np.array([[0.5, 0.9],
              [0.5, 0.1]])

pi = np.array([0.2, 0.8])

observations = ['S', 'C', 'C', 'C', 'S', 'C', 'S', 'C']

obs_map = {'S': 1, 'C': 0}  # 'S' = 1, 'C' = 0
observations_idx = [obs_map[o] for o in observations]

N = len(observations)

V = np.zeros((2, N))
backpointer = np.zeros((2, N), dtype=int)

V[:, 0] = pi * E[:, observations_idx[0]]

for t in range(1, N):
    for s in range(2):
        prob = V[:, t-1] * T[:, s] * E[s, observations_idx[t]]
        V[s, t] = np.max(prob)
        backpointer[s, t] = np.argmax(prob)

best_path = np.zeros(N, dtype=int)
best_path[N-1] = np.argmax(V[:, N-1])

for t in range(N-2, -1, -1):
    best_path[t] = backpointer[best_path[t+1], t+1]

state_map = {0: 'J', 1: 'B'}
best_path_states = [state_map[state] for state in best_path]

print("Secuencia observada:", observations)
print("Secuencia más probable de estados ocultos:", best_path_states)
print("Probabilidad de la secuencia más probable:", np.max(V[:, N-1]))

P_o = np.sum(V[:, N-1])
print("Probabilidad total de los estados observables P_o:", P_o)

norm_P_o = np.sum(np.exp(V[:, N-1]))
print("Verificación: La suma de los estados observables es 1 (después de normalización):", np.isclose(norm_P_o, 1.0))

pi_changed = np.array([0.5, 0.5])
V_changed = np.zeros((2, N))
backpointer_changed = np.zeros((2, N), dtype=int)

V_changed[:, 0] = pi_changed * E[:, observations_idx[0]]

for t in range(1, N):
    for s in range(2):
        prob = V_changed[:, t-1] * T[:, s] * E[s, observations_idx[t]]
        V_changed[s, t] = np.max(prob)
        backpointer_changed[s, t] = np.argmax(prob)

best_path_changed = np.zeros(N, dtype=int)
best_path_changed[N-1] = np.argmax(V_changed[:, N-1])

for t in range(N-2, -1, -1):
    best_path_changed[t] = backpointer_changed[best_path_changed[t+1], t+1]

best_path_states_changed = [state_map[state] for state in best_path_changed]

print("\nCon nueva probabilidad a-priori (0.5, 0.5):")
print("Secuencia más probable de estados ocultos:", best_path_states_changed)
print("Probabilidad de la secuencia más probable:", np.max(V_changed[:, N-1]))

P_o_changed = np.sum(V_changed[:, N-1])
print("Probabilidad total de los estados observables P_o (con nueva probabilidad a-priori):", P_o_changed)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

data = np.array([75, 80, 85, 70, 90, 95, 80, 85, 88, 92, 78, 84, 91, 76, 89, 95, 82, 79, 93, 86, 94, 87, 80, 81])

media = np.mean(data)
print(f"Media de las notas: {media}")

mediana = np.percentile(data, 50)
print(f"Mediana de las notas: {mediana}")

hist, bins = np.histogram(data, bins=10, density=True)
cdf = np.cumsum(hist) * (bins[1] - bins[0])

plt.figure(figsize=(10, 6))
plt.hist(data, bins=10, density=True, alpha=0.6, color='g', label="Histograma")
plt.plot(bins[1:], cdf, marker='o', linestyle='-', color='r', label="Distribución acumulada")
plt.title("Distribución acumulada de las notas")
plt.xlabel("Nota")
plt.ylabel("Probabilidad acumulada")
plt.legend()
plt.grid(True)
plt.show()

n_bootstrap = 1000
bootstrap_means = []

for _ in range(n_bootstrap):
    sample = np.random.choice(data, size=len(data), replace=True)
    bootstrap_means.append(np.mean(sample))

bootstrap_means = np.array(bootstrap_means)

plt.figure(figsize=(10, 6))
plt.hist(bootstrap_means, bins=30, density=True, alpha=0.6, color='b', label="Distribución de la media")
plt.title("Distribución de la media (bootstrapping)")
plt.xlabel("Media de la muestra")
plt.ylabel("Frecuencia")
plt.legend()
plt.grid(True)
plt.show()

mu, std = norm.fit(bootstrap_means)
print(f"Media ajustada (gaussiano): {mu}")
print(f"Desviación estándar ajustada (gaussiano): {std}")

plt.figure(figsize=(10, 6))
plt.hist(bootstrap_means, bins=30, density=True, alpha=0.6, color='b', label="Distribución de la media")
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2, label="Ajuste Gaussiano")
plt.title("Distribución de la media con ajuste gaussiano")
plt.xlabel("Media de la muestra")
plt.ylabel("Frecuencia")
plt.legend()
plt.grid(True)
plt.show()

mediana_bootstrap = np.percentile(bootstrap_means, 50)
print(f"Mediana de la distribución bootstrapped: {mediana_bootstrap}")

print("\nComparación de estimaciones:")
print(f"Media de las notas: {media}")
print(f"Mediana de las notas: {mediana}")
print(f"Media ajustada (gaussiano): {mu}")
print(f"Desviación estándar ajustada (gaussiano): {std}")
print(f"Mediana de la distribución bootstrapped: {mediana_bootstrap}")
```

```python
import numpy as np
from scipy.stats import gamma, norm, uniform

N = 10**4
a = np.array([1, 2, -1])

X1 = gamma.rvs(2, scale=3, size=N)
X2 = norm.rvs(5, np.sqrt(2), size=N)
X3 = uniform.rvs(0, 10, size=N)

X = a[0]*X1 + a[1]*X2 + a[2]*X3

mean_X = np.mean(X)
var_X = np.var(X)

mean_X_general = a[0]*np.mean(X1) + a[1]*np.mean(X2) + a[2]*np.mean(X3)
cov_X = np.array([[np.var(X1), np.cov(X1, X2)[0,1], np.cov(X1, X3)[0,1]],
                  [np.cov(X2, X1)[0,1], np.var(X2), np.cov(X2, X3)[0,1]],
                  [np.cov(X3, X1)[0,1], np.cov(X3, X2)[0,1], np.var(X3)]])
var_X_general = a.T @ cov_X @ a

correlation_X1_X2 = np.corrcoef(X1, X2)[0, 1]
correlation_X1_X3 = np.corrcoef(X1, X3)[0, 1]
correlation_X2_X3 = np.corrcoef(X2, X3)[0, 1]

def variance_of_mean(N, variables):
    cov_matrix = np.cov(variables, rowvar=False)
    var_mean = (1 / N**2) * np.sum(np.diagonal(cov_matrix)) + (2 / N**2) * np.sum(cov_matrix[np.triu_indices(len(cov_matrix), 1)])
    return var_mean

variables = np.vstack([X1, X2, X3]).T
var_mean_N2 = variance_of_mean(2, variables)

print("Resultados utilizando el array X:")
print(f"Media de X: {mean_X}")
print(f"Varianza de X: {var_X}")

print("\nResultados utilizando las definiciones generales:")
print(f"Media de X: {mean_X_general}")
print(f"Varianza de X: {var_X_general}")

print("\nCorrelaciones de Pearson:")
print(f"Correlación entre X1 y X2: {correlation_X1_X2}")
print(f"Correlación entre X1 y X3: {correlation_X1_X3}")
print(f"Correlación entre X2 y X3: {correlation_X2_X3}")

print(f"\nVarianza de la media para N=2: {var_mean_N2}")
```

```python
import numpy as np

def generar_puntos_sobre_esfera(N):

    phi = np.random.uniform(0, 2 * np.pi, N)
    costheta = np.random.uniform(-1, 1, N)
    theta = np.arccos(costheta)
    r = 1

    x = r * np.sin(theta) * np.cos(phi)
    y = r * np.sin(theta) * np.sin(phi)
    z = r * np.cos(theta)

    return x, y, z

def momentos_inercia(x, y, z):
    N = len(x)

    Ixx = np.mean(y**2 + z**2)

    Iyy = np.mean(x**2 + z**2)

    Izz = np.mean(x**2 + y**2)

    Ixy = -np.mean(x * y)

    return Ixx, Iyy, Izz, Ixy

N = 10**4

x, y, z = generar_puntos_sobre_esfera(N)

Ixx, Iyy, Izz, Ixy = momentos_inercia(x, y, z)

print(f"Momento de inercia Ixx: {Ixx}")
print(f"Momento de inercia Iyy: {Iyy}")
print(f"Momento de inercia Izz: {Izz}")
print(f"Producto de inercia Ixy: {Ixy}")

valor_exacto = 2 / 5
print(f"\nComparación con el valor exacto (2/5):")
print(f"Error relativo en Ixx: {abs(Ixx - valor_exacto)/valor_exacto}")
print(f"Error relativo en Iyy: {abs(Iyy - valor_exacto)/valor_exacto}")
print(f"Error relativo en Izz: {abs(Izz - valor_exacto)/valor_exacto}")
print(f"Error relativo en Ixy: {abs(Ixy)/valor_exacto}")
```

```python
import numpy as np
import scipy.special as sp
import matplotlib.pyplot as plt

alpha = 2
beta = 4

def f_beta(x, alpha, beta):
    return (sp.gamma(alpha + beta) / (sp.gamma(alpha) * sp.gamma(beta))) * (x**(alpha - 1)) * ((1 - x)**(beta - 1))

def metodo_aceptacion_rechazo(f, alpha, beta, num_muestras, incertidumbre):

    M = sp.gamma(alpha + beta) / (sp.gamma(alpha) * sp.gamma(beta))

    aceptadas = []
    muestras = 0
    while len(aceptadas) < num_muestras:

        x = np.random.uniform(0, 1)

        u = np.random.uniform(0, 1)

        if u <= f(x, alpha, beta) / (M * 1):
            aceptadas.append(x)

        muestras += 1

    area_estimado = np.mean([f(x, alpha, beta) for x in aceptadas])

    error_relativo = abs(area_estimado - 1)
    if error_relativo < incertidumbre:
        print(f"Estimación alcanzada con incertidumbre del {incertidumbre * 100}%")

    return area_estimado, aceptadas, muestras

num_muestras = 10000
incertidumbre = 0.01

area_estimado, aceptadas, muestras = metodo_aceptacion_rechazo(f_beta, alpha, beta, num_muestras, incertidumbre)

print(f"Área estimada bajo la curva de la Beta(2,4): {area_estimado}")
print(f"Número de muestras aceptadas: {len(aceptadas)}")
print(f"Número de muestras generadas: {muestras}")

x_vals = np.linspace(0, 1, 1000)
y_vals = f_beta(x_vals, alpha, beta)

plt.figure(figsize=(8, 6))

plt.plot(x_vals, y_vals, label='Distribución Beta(2,4)', color='blue', linewidth=2)

plt.hist(aceptadas, bins=50, density=True, alpha=0.6, color='orange', label='Muestras aceptadas')

plt.title('Método de Aceptación y Rechazo para Beta(2,4)')
plt.xlabel('x')
plt.ylabel('Densidad')
plt.legend()

plt.show()
```
